```python
import random
import numpy as np
import pandas as pd

with open("names.txt",encoding="utf-8") as file:
    content = file.read()

names_list = content.split()
names = random.choices(names_list,k=50)

with open("names.txt",encoding="utf-8") as file:
    content = file.read()
    
scores = np.random.randint(50,101,size=(50,5))
student_df = pd.DataFrame(scores,
             columns=['國文','英文','數學','地理','歷史'],
             index=names)
student_df
```

```python
from pandas import Series,DataFrame
def fun3(s:Series):
    sum=s.sum()
    mean=s.mean()
    count=s[s<60].count()
    Values=s[s<60].index.values
    return Series([sum,mean,count,Values],index=['總分','平均','不及格數','不及格的科目'])
student_df.apply(fun3,axis=1)
```
