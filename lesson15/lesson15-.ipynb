```python
import matplotlib.pyplot as plt

student1={'chinese':50,
          'math':20,
          'english':80,
          'history':100,
          'discover':60}
student2={'chinese':80,
          'math':29,
          'english':40,
          'history':90,
          'discover':60}
subjects=list(student1.keys())#key
score1=list(student1.values())#key
figure=plt.figure(figsize=(8,5))#
axes=figure.add_subplot(1,1,1)#
axes.plot(subjects,score1,'ro--',label='Student1')# 
score2=list(student2.values())#key
axes.plot(subjects,score2,'bo--',label='Student2')
axes.set_title('Table Of Student')#
axes.set_xlabel('Subjects')#
axes.set_ylabel('Score')#
axes.grid(axis='y')#
axes.legend()#


x_location=axes.get_xticks()
for i in x_location:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))



plt.show()#
```
