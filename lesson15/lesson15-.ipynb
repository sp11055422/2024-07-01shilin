```python
import matplotlib.pyplot as plt

student1={'chinese':50,
          'math':20,
          'english':80,
          'history':100,
          'discover':60}
student2={'chinese':80,
          'math':29,
          'english':40,
          'history':90,
          'discover':60}
subjects=list(student1.keys())#取出key
score1=list(student1.values())#取出key的值
figure=plt.figure(figsize=(8,5))#製造畫布
axes=figure.add_subplot(1,1,1)#製造圖表
axes.plot(subjects,score1,'ro--',label='Student1')#畫圖 
score2=list(student2.values())#取出key的值
axes.plot(subjects,score2,'bo--',label='Student2')
axes.set_title('Table Of Student')#標題
axes.set_xlabel('Subjects')#提示
axes.set_ylabel('Score')#提示
axes.grid(axis='y')#輔助格線
axes.legend()#提示線的名子


x_location=axes.get_xticks()
for i in x_location:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))



plt.show()#顯示畫的圖
```
